### Recursive function:

    -A function whihc is called by itself,is known as recursive function invocation.
    -pyhton supports recursive function invocation,but it will not support infinite recursion.

```python
# infinite recursion
def myfunc():
    print('welcome')
    myfunc()
myfunc()
```

```python
# finite recursion
i=1
def myfunc():
    print('welcome')
    global i
    i=i+1
    while i<=5:
        myfunc()
myfunc()
```

```python
def recur_fact(x):
    if x==1:
        return 1
    else:
        return (x * recur_fact(x-1))
num=int(input('Enter a number:'))
if num >=1:
    print('The factorial of' ,num, 'is',recur_fact(num))
```

### Lambda fucntion:

    -A function ,which doesn't contain any name is know as a Anonymous function or Lambda function.
    
    -We can not call the lambda function,directly because it doesn't contain any name.
    
    -We can assign the lambda function to a variable and ,we can call that Lambda function ,through that variable.
    
    -Whenever,we call a Lambda function ,through a variable,it will return the expression value.


```python
lambda var1,var2,varn: Expression
```

```python
print('start')
myfunc=lambda x,y:x*y
p=myfunc(20,30)
print(p)
print('end')
```

#### The above lambda function is equal to the following code.

```python
print('start')
def myfunc(x,y):
    return x*y
p=myfunc(20,30)
print(p)
print('end')
```

```python
#passing one function value to another
def cube(a,function):
    p=function(a)
    return a*p
def square(b): 
    return b*b
x=cube(12,square)
print(x)
```

#### Wheneve we want to pass ,one function as a parameter to the another function,then we use lambda function.

```python
#passing one function value to another,using 'LAMBDA' function
def cube(a,function):
    p=function(a)
    return a*p
x=cube(13,lambda b:b*b)
print(x)
```

### Filter

    -filter function is used to check the condition

```python
my_list=[1,3,4,6,8,9,11,12]
new_list=list(filter(lambda x:x%2==0,my_list))
print(new_list)
```

### Map
    -map function is used to evalutaion through given condition.

```python
my_list=[1,3,4,6,8,9,11,12]
new_list=list(map(lambda x:x*x,my_list))
print(new_list)
```
